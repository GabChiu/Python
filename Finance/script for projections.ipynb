```python
x = [20,12.3,8.84,7.74,22]
y = [63.74,18.22,10.74,17.51,43.06]
sns.set()
n = ["CDAY", "ADP", "ORCL", "SAP", "WDAY"]


m, b = np.polyfit(x, y, 1)




fig, ax = plt.subplots()
ax.plot(x, y)
plt.plot(x, m*x + b)

plt.xlabel("Growth rate (%)")
plt.ylabel("EV/EBITDA forward 12M")

for i, txt in enumerate(n):
    ax.annotate(txt, (x[i], y[i]))
    
```

```python
barWidth = 0.2
strategy = [0.285,0.138,0.084,0.052,0.027]
casual=[0.309,0.151,0.09,0.054,0.028]
hyperCasual=[0.327,0.157,0.088,0.05,0.023]
socialCasino=[0.322,0.171,0.11,0.072,0.042]

r1=np.arange(len(strategy))
r2=[x+barWidth for x in r1]
r3=[x+barWidth for x in r2]
r4=[x+barWidth for x in r3]
```

```python
sns.set()
plt.figure(figsize=(10,5))
plt.bar(r1, strategy, color="darkblue", width=barWidth, edgecolor="white", label="Strategy")
plt.bar(r2, casual, color="gold", width=barWidth, edgecolor="white", label="Casual")
plt.bar(r3, hyperCasual, color="black", width=barWidth, edgecolor="white", label="HyperCasual")
plt.bar(r4, socialCasino, color="wheat", width=barWidth, edgecolor="white", label="SocialCasino")

plt.xticks([r+barWidth for r in range(len(strategy))], ["Day 1", "Day 3", "Day 7", "Day 14", "Day 30"])
plt.legend()
plt.title("Retention Rate")
plt.show()
```
