```python
import numpy as np
import pandas as pd
from pandas_datareader import data
from matplotlib import pyplot as plt
from datetime import date
```

```python
start = date(date.today().year-5, date.today().month, date.today().day)
end = date.today()
# Big 6
rbc = data.DataReader("RY.TO", start=start, end=end, data_source="yahoo")
td = data.DataReader("TD.TO", start=start, end=end, data_source="yahoo")
scotia = data.DataReader("BNS.TO", start=start, end=end, data_source="yahoo")
cibc = data.DataReader("CM.TO", start=start, end=end, data_source="yahoo")
bmo = data.DataReader("BMO.TO", start=start, end=end, data_source="yahoo")
nbc = data.DataReader("NA.TO", start=start, end=end, data_source="yahoo")
```

```python
plt.figure(figsize=(5,5))
rbc["Adj Close"].plot()
plt.figure(figsize=(5,5))
rbc.Volume.plot()
```

```python
# Yearly average number of shares outstanding for MSFT, AAPL
rbc_shares = {2015:1443e6, 2016:1485e6, 2017:1458e6, 2018: 1441e6, 2019: 1433e6, 2020: 1424e6}
td_shares = {2015:1854e6, 2016:1856e6, 2017:1849e6, 2018: 1829e6, 2019: 1820e6, 2020: 1814e6}
scotia_shares = {2015:1203e6, 2016:1206e6, 2017:1198e6, 2018: 1228e6, 2019: 1216e6, 2020: 1211e6}
cibc_shares = {2015:397e6, 2016:395e6, 2017:440e6, 2018: 442e6, 2019: 445e6, 2020: 446e6}
bmo_shares = {2015:642e6, 2016:646e6, 2017:648e6, 2018: 640e6, 2019: 639e6, 2020: 643e6}
nbc_shares = {2015:337e6, 2016:338e6, 2017:341e6, 2018: 337e6, 2019: 335e6, 2020: 336e6}
```

```python
bank_shares = dict()
bank_shares["rbc"] = {2015:1443e6, 2016:1485e6, 2017:1458e6, 2018: 1441e6, 2019: 1433e6, 2020: 1424e6}
bank_shares["td"] = {2015:1854e6, 2016:1856e6, 2017:1849e6, 2018: 1829e6, 2019: 1820e6, 2020: 1814e6}
bank_shares["scotia"] = {2015:1203e6, 2016:1206e6, 2017:1198e6, 2018: 1228e6, 2019: 1216e6, 2020: 1211e6}
bank_shares["cibc"] = {2015:397e6, 2016:395e6, 2017:440e6, 2018: 442e6, 2019: 445e6, 2020: 446e6}
bank_shares["bmo"] = {2015:642e6, 2016:646e6, 2017:648e6, 2018: 640e6, 2019: 639e6, 2020: 643e6}
bank_shares["nbc"] = {2015:337e6, 2016:338e6, 2017:341e6, 2018: 337e6, 2019: 335e6, 2020: 336e6}
```

```python
rbc["Year"]=rbc.index.year
rbc.reset_index(level=0, inplace=True)
scotia["Year"]=scotia.index.year
scotia.reset_index(level=0, inplace=True)
td["Year"]=td.index.year
td.reset_index(level=0, inplace=True)
cibc["Year"]=cibc.index.year
cibc.reset_index(level=0, inplace=True)
bmo["Year"]=bmo.index.year
bmo.reset_index(level=0, inplace=True)
nbc["Year"]=nbc.index.year
nbc.reset_index(level=0, inplace=True)
```

```python
def add_market_cap(bank, bank_shares):
    bank = bank.copy(deep=True)
    bank["cap"]=0

    for i, year in enumerate(bank["Year"]):
        shares = bank_shares.get(year)
        bank.cap.iloc[i] = shares * bank["Adj Close"].iloc[i]
    return bank
```

```python
rbc = add_market_cap(rbc, rbc_shares)
scotia = add_market_cap(scotia, scotia_shares)
td = add_market_cap(td, td_shares)
cibc = add_market_cap(cibc, cibc_shares)
bmo = add_market_cap(bmo, bmo_shares)
nbc = add_market_cap(nbc, nbc_shares)
```

```python
plt.figure(figsize=(10,7))
plt.plot(rbc.Date, rbc.cap, label="RY", color="blue")
plt.plot(scotia.Date, scotia.cap, label="BNS", color="red")
plt.plot(td.Date, td.cap, label="TD", color="green")
plt.plot(cibc.Date, cibc.cap, label="CM", color="orange")
plt.plot(bmo.Date, bmo.cap, label="BMO", color="purple")
plt.plot(nbc.Date, nbc.cap, label="NA", color="crimson")
plt.legend()
plt.title("Market Cap")
```

```python
import fbprophet
from fbprophet.plot import add_changepoints_to_plot
```

```python
rbc = rbc.rename(columns={"Date":"ds","cap":"y"})
rbc.y = rbc.y/1e9
```

```python
rbc_prophet = fbprophet.Prophet(changepoint_prior_scale=0.15, yearly_seasonality=True)
rbc_prophet.fit(rbc)
rbc_forecast=rbc_prophet.make_future_dataframe(periods=365,freq="D")
rbc_forecast=rbc_prophet.predict(rbc_forecast)
fig = rbc_prophet.plot(rbc_forecast, xlabel="Date", ylabel="Market Cap")
ad = add_changepoints_to_plot(fig.gca(), rbc_prophet, rbc_forecast)
```

```python
rbc_prophet = fbprophet.Prophet(changepoint_prior_scale=0.15, yearly_seasonality=True, changepoint_range=0.85)
rbc_prophet.fit(rbc)
rbc_forecast=rbc_prophet.make_future_dataframe(periods=365,freq="D")
rbc_forecast=rbc_prophet.predict(rbc_forecast)
fig = rbc_prophet.plot(rbc_forecast, xlabel="Date", ylabel="Market Cap")
ad = add_changepoints_to_plot(fig.gca(), rbc_prophet, rbc_forecast)
```

Changepoints represents when the time series growth rate significantly changes. Changepoints are limited to the first 80% of the time series by default. Facebook Prophet's description of changepoints here: https://facebook.github.io/prophet/docs/trend_changepoints.html

The changepoint_prior_scale reflects how closely the trend tracks the given data (black points). The higher, the more trend flexibility and change of overfit.
