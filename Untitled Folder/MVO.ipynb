```python
import numpy as np
import pandas as pd
from pandas_datareader import data
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import date
```

```python
sp500 = pd.read_csv("sp500.csv")
tickers = sp500["Symbol"]
today = date.today()
```

```python
test_tickers = tickers[0:5]
```

```python
tickers_data = data.DataReader(test_tickers,start="2018-1-1", end = today, data_source="yahoo")["Adj Close"]
```

```python
returns = tickers_data.pct_change() # pct change between current and prior row
cumulative_returns = ((1 + returns).cumprod()-1)
```

```python
returns.plot(figsize = (20,10))
```

```python
expected_returns = returns.mean()
cov_matrix = returns.cov()
```

Efficient frontier

```python
np.random.seed(2020)
```

```python
def portfolio_performance(weights, expected_returns, cov_matrix):
    # expected returns and cov_matrix variables are expected to be daily
    # the below annualizes the values
    pf_return = np.sum(weights*expected_returns)*252
    pf_std = np.sqrt(np.dot(weights.T,np.dot(cov_matrix, weights)))*np.sqrt(252)
    return pf_return, pf_std

def random_portfolio(expected_returns, cov_matrix, risk_free = 0.01):
    # initialize every security with a randomized weighting
    weights = np.random.random(expected_returns.count())
    weights /= np.sum(weights)
    
    pf_return, pf_std = portfolio_performance(weights, expected_returns, cov_matrix)
    pf_sharpe = (pf_return - risk_free)/pf_std
    return weights, pf_return, pf_std, pf_sharpe
```

```python
sample = 10000

stats = np.zeros((3,sample))
for i in range(sample):
    weights, pf_return, pf_std, pf_sharpe = random_portfolio(expected_returns, cov_matrix)
    stats[0,i] = pf_return
    stats[1,i] = pf_std
    stats[2,i] = pf_sharpe
```
