```python
import string
import collections
from pprint import pprint
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
```

```python
def process_text(text, stem=True):
    """ Tokenize text and stem words removing punctuation """
    text = text.translate(str.maketrans(string.punctuation, " "*len(string.punctuation))
                         ).replace(" "*4, " ").replace(" "*3," ").replace(" "*2," ").strip()
    tokens = word_tokenize(text)
 
    if stem:
        stemmer = PorterStemmer()
        tokens = [stemmer.stem(t) for t in tokens]
 
    return tokens
 

def cluster_texts(texts, clusters=3):
    """ Transform texts to Tf-Idf coordinates and cluster texts using K-Means """
    vectorizer = TfidfVectorizer(tokenizer=process_text,
                                 stop_words=stopwords.words('english'),
                                 max_df=0.3,
                                 min_df=0.01,
                                 lowercase=True)
    tfidf_model = vectorizer.fit_transform(texts)
    km_model = KMeans(n_clusters=clusters)
    km_model.fit(tfidf_model)
    
    print("Top words per Cluster")
    order_centroids = km_model.cluster_centers_.argsort()[:,::-1]
    terms = vectorizer.get_feature_names()
    for i in range(clusters):
        print("Cluster %d:" %i),
        for ind in order_centroids[i, :10]:
            print(" %s" % terms[ind]),
        print
    
    clustering = collections.defaultdict(list)
 
    for idx, label in enumerate(km_model.labels_):
        clustering[label].append(idx)
 
    # plot the clusters
    kmean_indices = km_model.fit_predict(tfidf_model)
    pca = PCA(n_components=2)
    scatter_plot_points = pca.fit_transform(tfidf_model.toarray())
    colors = ["r", "b", "c", "y", "m", "bisque", "lime", "lightsteelblue", "darkgray", "mistyrose", 
              "darkkhaki","paleturquoise","violet" ]
    x_axis = [o[0] for o in scatter_plot_points]
    y_axis = [o[1] for o in scatter_plot_points]
    fig, ax = plt.subplots(figsize=(20,10))

    ax.scatter(x_axis, y_axis, c=[colors[d] for d in kmean_indices])
    
    # calculate distortion
    distortion = []
    K = range(1,10)
    for k in K:
        kModel = KMeans(n_clusters=k)
        kModel.fit(tfidf_model)
        distortion.append(kModel.inertia_)
    plt.figure(figsize=(10,5))
    plt.plot(K, distortion, "bx-")
    plt.title("Elbow Graph")


    return clustering
```

```python
articles = data.content.dropna().to_list()
```

```python
clusters = cluster_texts(articles, 4)
```

```python
results = dict(clusters)
```

```python
def identify_author(cluster, author):
    df = pd.DataFrame()
    for key in cluster:
        temp = []
        for clause in cluster[key]:
            temp.append(author[clause])
        df.insert(0, key, pd.Series(temp))
    return df
```

```python
allocated_authors = identify_author(results, data.author)
```
