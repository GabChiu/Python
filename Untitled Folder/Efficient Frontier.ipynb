```python
import numpy as np
import pandas as pd
from pandas_datareader import data
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import date
```

```python
sp500 = pd.read_csv("sp500.csv")
tickers = sp500["Symbol"]
today = date.today()
```

```python
test_tickers = tickers[100:125]
```

```python
tickers_data = data.DataReader(test_tickers,start="2018-1-1", end = today, data_source="yahoo")["Adj Close"]
```

```python
returns = tickers_data.pct_change() # pct change between current and prior row
```

```python
returns.plot(figsize = (20,10))
cumulative_returns = ((1 + returns).cumprod()-1)
```

```python
expected_returns = returns.mean()
cov_matrix = returns.cov()
```

Efficient frontier with large number of assets

```python
np.random.seed(2020)
```

```python
def portfolio_performance(weights, expected_returns, cov_matrix):
    # expected returns and cov_matrix variables are expected to be daily
    # the below annualizes the values
    pf_return = np.sum(weights*expected_returns)*252
    pf_std = np.sqrt(np.dot(weights.T,np.dot(cov_matrix, weights)))*np.sqrt(252)
    return pf_return, pf_std

def random_portfolio(expected_returns, cov_matrix, number, risk_free = 0.01):
    # initialize every security with a randomized weighting

    stats = np.zeros((3, number))
    weights_stats = []
    
    for i in range(number):
        weights = np.random.random(expected_returns.count())
        weights /= np.sum(weights)
        weights_stats.append(weights)
        
        pf_return, pf_std = portfolio_performance(weights, expected_returns, cov_matrix)
        pf_sharpe = (pf_return - risk_free)/pf_std
        stats[0,i] = pf_std
        stats[1,i] = pf_return
        stats[2,i] = pf_sharpe    
    
    return stats, weights_stats
```

```python
def plot_frontier(expected_returns, cov_matrix, risk_free=0.01, samples=50000):
    stats, weights = random_portfolio(expected_returns, cov_matrix, samples, risk_free)
    
    max_sharpe = np.argmax(stats[2])
    std_max_sharpe, r_max_sharpe = stats[0, max_sharpe], stats[1, max_sharpe]

    max_sharpe_weights = pd.DataFrame(weights[max_sharpe], index=expected_returns.index, columns=['allocation'])
    max_sharpe_weights.allocation = [round(i*100,2) for i in max_sharpe_weights.allocation]
    max_sharpe_weights = max_sharpe_weights.T
    
    print("Max Sharpe Ratio Portfolio Allocation\n")
    print("Annualized Return: ", round(r_max_sharpe,2))
    print("Annualized Volatility: ", round(std_max_sharpe,2))
    print("\n")
    print(max_sharpe_weights)
    
    plt.figure(figsize=(10,7))
    plt.scatter(stats[0,:],stats[1,:],c=stats[2,:], marker="o", cmap='plasma', s=20, alpha=0.3)
    plt.colorbar()
    plt.xlabel('Annualized volatility')
    plt.ylabel('Annualized return')
    
```

```python
plot_frontier(expected_returns, cov_matrix)
```

```python
plot_frontier(expected_returns, cov_matrix, samples=1000000)
```

Efficient frontier Max sharpe ratio portfolio and minimum variance portfolio

```python
def eff_frontier(expected_returns, cov_matrix, risk_free=0.01, samples=50000):
    stats, weights = random_portfolio(expected_returns, cov_matrix, samples, risk_free)
    
    max_sharpe = np.argmax(stats[2])
    std_max_sharpe, r_max_sharpe = stats[0, max_sharpe], stats[1, max_sharpe]

    max_sharpe_weights = pd.DataFrame(weights[max_sharpe], index=expected_returns.index, columns=['allocation'])
    max_sharpe_weights.allocation = [round(i*100,2) for i in max_sharpe_weights.allocation]
    max_sharpe_weights = max_sharpe_weights.T
    
    min_var = np.argmin(stats[0])
    std_min_var, r_min_var = stats[0, min_var], stats[1, min_var]
    min_var_weights = pd.DataFrame(weights[min_var], index=expected_returns.index, columns=['allocation'])
    min_var_weights.allocation = [round(i*100,2) for i in min_var_weights.allocation]
    min_var_weights = min_var_weights.T
    
    print("Max Sharpe Ratio Portfolio Allocation\n")
    print("Annualized Return: ", round(r_max_sharpe,2))
    print("Annualized Volatility: ", round(std_max_sharpe,2))
    print("\n")
    print(max_sharpe_weights)
    
    print("Min Var Portfolio Allocation\n")
    print("Annualized Return: ", round(r_min_var,2))
    print("Annualized Volatility: ", round(std_min_var,2))
    print("\n")
    print(min_var_weights)

    # the efficient frontier is constructed as a linear combination of Max Sharpe and Min Var portfolios
    
    mv_weight = 0
    tick = 100
    stats = np.zeros((3, tick))
    
    # expected_returns = pd.Series([r_max_sharpe, r_min_var],index=["Max Sharpe", "Min Var"])
    for i in range(tick):
        ms_weight = 1 - mv_weight
        pf_weight = (ms_weight*max_sharpe_weights + mv_weight*min_var_weights)/100
        pf_weight = pf_weight.to_numpy()
        pf_weight = pf_weight.reshape(pf_weight.size)
        pf_return, pf_std = portfolio_performance(pf_weight, expected_returns, cov_matrix)
        stats[0,i] = pf_std
        stats[1,i] = pf_return
        stats[2,i] = (pf_return - risk_free)/pf_std
        mv_weight += 1/tick
        
        
    plt.figure(figsize=(10,7))
    plt.scatter(stats[0,:],stats[1,:],c=stats[2,:], marker="o", cmap='plasma', s=20, alpha=0.3)
    plt.colorbar()
    plt.scatter(std_max_sharpe, r_max_sharpe, marker="*", color="b", s=100, label="Max Sharpe")
    plt.scatter(std_min_var, r_min_var, marker="*", color="r", s=100, label = "Min Var")
    plt.xlabel('Annualized volatility')
    plt.ylabel('Annualized return')
    
```

```python
eff_frontier(expected_returns, cov_matrix, samples = 1000)
```
