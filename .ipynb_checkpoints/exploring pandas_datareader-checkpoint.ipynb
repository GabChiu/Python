```python
from datetime import date
start = date(date.today().year-4, date.today().month, date.today().day)
end = date.today()
tsx = pdr.DataReader("^GSPTSE", start=start, end=end, data_source="yahoo")
```

```python
tsx["Adj Close"].plot()
```

```python
(tsx["Adj Close"]/tsx["Adj Close"][0]).plot()
```

```python
tsx["Adj Close"].pct_change().plot()
```

```python
tsxReturns = tsx.pct_change(periods=60)
tsxReturns["Adj Close"].plot()
```

```python
tsx.pct_change().dropna()["Adj Close"].hist()
```
