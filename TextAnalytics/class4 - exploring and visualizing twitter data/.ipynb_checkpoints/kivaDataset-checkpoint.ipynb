```python
import pandas as pd

kiva_df=pd.read_csv("kiva_cleaned.csv")
kiva_df.head(5)
```

```python
kiva_df.dropna(axis=0,inplace=True)
kiva_df.shape
```

```python
kiva_df.head()
kiva_df.tail()
```

```python
kiva_df.reset_index(drop=True, inplace=True)
```

```python
kiva_df.tail()
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
import unidecode
```

```python
stop_words = set(stopwords.words("english") + stopwords.words("spanish"))
lemmer = WordNetLemmatizer()

def preprocess(x):
    x=x.lower()
    x=re.sub(r"[^\w\s]","" ,x)
    x=unidecode.unidecode(x)
    x=re.sub(r"\d+","",x)
    x=[lemmer.lemmatize(w) for w in x.split() if w not in stop_words]
    return " " .join(x)
```

```python
kiva_df["en_clean_pre"] = kiva_df["en_clean"].apply(preprocess)
```

```python
kiva_df.head()
```

Sci-kit Learn

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_df=0.5, min_df=0.05, max_features=1000, ngram_range=[1,3])
dtm = vectorizer.fit_transform(kiva_df["en_clean_pre"])
dtm.shape
```
